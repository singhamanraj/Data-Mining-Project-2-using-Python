```python
#India Pollution on different states 

import math
import pandas as pd
import matplotlib.pyplot as pp
import plotly
from plotly import tools
import plotly.plotly as py
import plotly.graph_objs as go
py.sign_in('monicams89','Z9lw6RBF9YBbVmD8NUKC')
plotly.offline.init_notebook_mode()
```

```python
#Read Data excluding p_value column
col_names=['stn_code','sampling_date','State_location','agency','type','so2','no2','pm2_rspm','location_monitoring_station','date','Year']
indiapoll = pd.read_csv('India_Pollution.csv',encoding="cp1252", index_col=0,header=0,usecols=[0,1,2,3,4,6,7,8,9,10,11],skiprows=5,na_values="na",names=col_names)
indiapoll.head(10)
```

```python
#Dimension of dataset
indiapoll.shape
```

```python
#Remove Na and blank values
indiapoll.dropna(how="any",inplace=True)
indiapoll.shape
```

```python
#Split state_location column in 2 different attributes state and location

indiapoll[['State','location']] = indiapoll['State_location'].str.split(',',expand=True)
indiapoll.head(10)
```

```python
#Drop State_location column
indiapoll.drop('State_location',axis = 1, inplace = True)
```

```python
indiapoll.head(10)
```

```python
#Split pm2 and rspm column
indiapoll[['pm2','rspm']] = indiapoll['pm2_rspm'].str.split(',',expand=True)
indiapoll.head(10)
```

```python
#Split pm2 and rspm column and fetch only value
indiapoll['pm2'] = indiapoll['pm2'].str.split().str[-1]
indiapoll['rspm'] = indiapoll['rspm'].str.split().str[-1]
indiapoll.head(10)
```

```python
#Drop pm2_rspm column
indiapoll.drop('pm2_rspm',axis = 1, inplace = True)
```

```python
#Fetch Date and remove unwanted value
indiapoll['date'] = indiapoll['date'].str.split().str[0]
indiapoll.head(10)
```

```python
#Top 10 states emiting So2

top_SO2 = indiapoll.sort_values(by='so2', ascending= False) 
top_SO2 = top_SO2.head(10)
top_SO2
```

```python
#Top 10 states emiting NO2

top_NO2 = indiapoll.sort_values(by='no2', ascending= False) 
top_NO2 = top_NO2.head(10)
top_NO2
```

```python
#Converting rspm value to numeric value
indiapoll['rspm']=pd.to_numeric(indiapoll['rspm'])

```

```python
#Top 10 states emiting RSPM

top_rspm = indiapoll.sort_values(by='rspm', ascending= False) 
top_rspm = top_rspm.head(10)
top_rspm
```

```python
#Comparing Top 10 Emitters for SO2, NO2 and rspm 

trace1 = go.Bar(
    x=top_SO2['State'],
    y=top_SO2['so2'],
    name='so2'
)
trace2 = go.Bar(
    x=top_NO2['State'],
    y=top_NO2['no2'],
    name='no2'
)
trace3 = go.Bar(
    x=top_rspm['State'],
    y=top_rspm['rspm'],
    name='rspm'
)
data = [trace1, trace2,trace3]
layout = go.Layout(
    barmode='stack'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Bar-Plot')
```

```python
#Pollution levels(So2,No2,rspm) for a particular state
def state_pollution(state):
    indiapoll_indexed_state = indiapoll.set_index(['State']).sort_index()
    indiapoll_indexed_state.head()

    state_data = indiapoll_indexed_state.loc[state]
    state_so2= sum(state_data['so2'])
    state_no2= sum(state_data['no2'])
    state_rspm= sum(state_data['rspm'])

    return[state_so2,state_no2,state_rspm]
    
    
```

```python
state_name=state_pollution('Delhi')
```

```python
labels1 = ['so2','no2','rspm']
values1 = state_name

trace1 = go.Pie(labels=labels1, values=values1,pull=.2,hole=.2)


py.iplot([trace1], filename='basic_pie_chart')
```

```python
# Subplots to show comparison for SO2 NO2 and rspm

trace1 = go.Scatter(
    x=top_NO2['State'],
    y=top_NO2['no2'],
    name = 'no2'
)
trace2 = go.Scatter(
    x=top_SO2['State'],
    y=top_SO2['so2'],
    name = 'so2'
)

trace3 = go.Scatter(
    x=top_rspm['State'],
    y=top_rspm['rspm'],
    name = 'rspm'
)
fig = tools.make_subplots(rows=1, cols=3)

fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 1, 2)
fig.append_trace(trace3, 1, 3)

fig['layout'].update(height=600, width=1000, title='Subplots Showing The Comparison')
py.iplot(fig, filename='multiple-subplots')
```

```python
# Examine pollution levels in type of areas i.e Residential vs Rural vs RIRUO

indiapoll_indexed = indiapoll.set_index(['type']).sort_index()
indiapoll_indexed.head()


indiapoll_type_industry = indiapoll_indexed.loc['Industrial Area']
Sum_industry_so2 = sum(indiapoll_type_industry['so2'])
Sum_industry_no2 = sum(indiapoll_type_industry['no2'])
Sum_industry_rspm = sum(indiapoll_type_industry['rspm'])

indiapoll_type_residential = indiapoll_indexed.loc['Residential, Rural and other Areas']
Sum_residential_so2 = sum(indiapoll_type_residential['so2'])
Sum_residential_no2 = sum(indiapoll_type_residential['no2'])
Sum_residential_rspm = sum(indiapoll_type_residential['rspm'])


indiapoll_type_riruo = indiapoll_indexed.loc['RIRUO']
Sum_riruo_so2 = sum(indiapoll_type_riruo['so2'])
Sum_riruo_no2 = sum(indiapoll_type_riruo['no2'])
Sum_riruo_rspm = sum(indiapoll_type_riruo['rspm'])

```

```python
#Ploting graph

trace11 = go.Bar(
    x=['Industrial Area','Residential, Rural and other Areas','RIRUO'],
    y=[Sum_industry_so2,Sum_residential_so2,Sum_riruo_so2],
    name='so2'
)
trace22 = go.Bar(
    #x=indiapoll['type'],
    x=['Industrial Area','Residential, Rural and other Areas','RIRUO'],
    y=[Sum_industry_no2,Sum_residential_no2,Sum_riruo_no2],
    name='no2'
)
trace33 = go.Bar(
    #x=indiapoll['type'],
    x=['Industrial Area','Residential, Rural and other Areas','RIRUO'],
    y=[Sum_industry_rspm,Sum_residential_rspm,Sum_riruo_rspm],
    name='rspm'
)

data11 = [trace11,trace22,trace33]
layout = go.Layout(
    barmode='bar'
)

fig = go.Figure(data=data11, layout=layout)
py.iplot(fig, filename='Bar-Plot')
```

```python
# Plot trend for 2014 and show comparison between States for North India and South India 

northIndia=['Haryana','Lucknow','Punjab']
southIndia=['Karnataka','Odisha', 'Tripura']

indiapoll_indexed_year = indiapoll.set_index(['Year']).sort_index()
indiapoll_indexed_year.head()

indiapoll_indexed_state2014=indiapoll_indexed_year.loc[2014].set_index(['State'])
indiapoll_indexed_state2015=indiapoll_indexed_year.loc[2015].set_index(['State'])

trace1=go.Scatter(
    x=northIndia,
    y=indiapoll_indexed_state2014.loc[northIndia]['so2'],
    fill='tonexty',
    name = 'North India - 2014'
)
trace2=go.Scatter(
    x=southIndia,
    y=indiapoll_indexed_state2014.loc[southIndia]['so2'],
    fill='tonexty',
    name = 'South India - 2014'
)

trace3=go.Scatter(
    x=northIndia,
    y=indiapoll_indexed_state2015.loc[northIndia]['so2'],
    fill='tonexty',
    name = 'North India - 2015'
)
trace4=go.Scatter(
    x=southIndia,
    y=indiapoll_indexed_state2015.loc[southIndia]['so2'],
    fill='tonexty',
    name = 'South India - 2015'
)

fig = tools.make_subplots(rows=2, cols=2)

fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 1, 2)
fig.append_trace(trace3, 2, 1)
fig.append_trace(trace4, 2, 2)


fig['layout'].update(height=600, width=1000, title='Subplots Showing The Comparison for North and South India (2014) vs (2015)')
py.iplot(fig, filename='multiple-subplots')
```
